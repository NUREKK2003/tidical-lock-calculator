### Config
### !IMPORTANT - We assume that planet doesn't have the moon

```python
Q_disp_factor = 100 # for Earth like planets
m_planet_mass = 1 # 1 - mass of the Earth
a_semi_distance = 1 # semiaxis distance from star in AU
M_bigger_mass = 1 # 1 - mass of the Sun
r_planet = 1 # 1 - radius of the earth
P_time = 6 # initial rotation peroid in hours (for the Earth it was 6 hours after collision with Thea)
```

```python
# Result in bilions years
def get_time_to_tidical_lock(Q_disp_factor = 100,m_planet_mass = 1,a_semi_distance = 1,M_bigger_mass = 1,r_planet = 1,P_time = 6):
    t = 24*((Q_disp_factor*m_planet_mass*pow(a_semi_distance,6))/(P_time*pow(M_bigger_mass,2)*pow(r_planet,3)))
    return t
```

```python
# Sun and the Earth
print(get_time_to_tidical_lock())
```

```python
import pandas as pd
import plotly.express as px
```

```python
results = []

for i in range(100,1000):
    distance = 1*(i/1000)
    t = get_time_to_tidical_lock(Q_disp_factor = 100,m_planet_mass = 1,a_semi_distance = distance,M_bigger_mass = 1,r_planet = 1,P_time = 6)
    results.append([distance,t])
results_df = pd.DataFrame(results,columns=['distance','t'])
```

```python
fig = px.line(results_df, x="distance", y="t", title='Tidical Lock Time vs Semi-major axis')
fig.update_layout(yaxis_type="log")
fig.update_layout(title_x=0.5)
fig.update_layout(
    xaxis_title="Distace (1 =  Distance from Earth to Sun)", yaxis_title="Time to tidical lock (in G years)"
)
fig.show()
```

```python
# Earth around Proxima Centauri

results = []

for i in range(100,1000):
    distance = 1*(i/10000)
    t = get_time_to_tidical_lock(Q_disp_factor = 100,m_planet_mass = 1,a_semi_distance = distance,M_bigger_mass = 0.12,r_planet = 1,P_time = 6)*1000
    results.append([distance,t])
results_df = pd.DataFrame(results,columns=['distance','t'])

fig = px.line(results_df, x="distance", y="t", title='Tidical Lock Time vs Semi-major axis')
fig.update_layout(yaxis_type="log")
fig.update_layout(title_x=0.5)
fig.update_layout(
    xaxis_title="Distace (1 =  Distance from Earth to Sun)", yaxis_title="Time to tidical lock (in M years)"
)
fig.add_vline(x=0.05, line_width=3, line_dash="dash", line_color="green")
fig.show()
```
